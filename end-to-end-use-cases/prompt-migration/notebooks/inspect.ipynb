```python
import pandas as pd
from tabulate import tabulate
from IPython.display import display

def display_incorrect_answers(csv_path):
    df = pd.read_csv(csv_path)
    
    incorrect_answers = df[df['Predicted Answer'] != df['Correct Answer']]
    
    max_rows = 300  # Adjust as needed
    incorrect_answers_limited = incorrect_answers[['Question', 'Predicted Answer', 'Correct Answer']].head(max_rows)
    
    display(pd.DataFrame(incorrect_answers_limited))
    
    if len(incorrect_answers) > max_rows:
        print(f"\nShowing first {max_rows} rows out of {len(incorrect_answers)} incorrect answers.")
    
    total_questions = len(df)
    incorrect_count = len(incorrect_answers)
    accuracy = ((total_questions - incorrect_count) / total_questions) * 100
    print(f"\nAccuracy: {accuracy:.2f}%")

display_incorrect_answers(csv_path)
```

```python
import pandas as pd
from IPython.display import display

def check_correct_answers(csv_path):
    df = pd.read_csv(csv_path)
    
    correct_mismatches = df[(df['Is Correct'] == True) & (df['Predicted Answer'] != df['Correct Answer'])]
    
    if not correct_mismatches.empty:
        print("\nWarning: Some 'Is Correct' rows have mismatched answers!")
        display(correct_mismatches[['Question', 'Predicted Answer', 'Correct Answer', 'Is Correct']])
    else:
        print("\nAll 'Is Correct' rows have matching answers.")

 
check_correct_answers(csv_path)
```
